```python
import regex as re
```

```python
def hindi_preprocessing(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '',text)
    words = text.split()
    words = [word for word in words if word not in hindi_stopwords]

    return ' '.join(words)
```

```python
df['Hindi Lyrics'] = df['Hindi Lyrics'].apply(lambda x: hindi_preprocessing(x))
```

```python
df
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmtizer = WordNetLemmatizer()
def english_preprocessing(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '',text)
    words = text.split()
    words = [word for word in words if word not in english_stopwords]
    words = [lemmtizer.lemmatize(word) for word in words]
    
    return ' '.join(words)
```

```python
df['English Translations'] = df['English Translations'].apply(lambda x: english_preprocessing(x))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=8000, ngram_range=(1,2))
```

```python
tfidf_matrix = tfidf.fit_transform(df['Hinglish Lyrics'])
```

```python
tfidf_matrix.shape
```

```python
df = df.drop_duplicates(subset='Song Title')
```

```python
df = df.reset_index(drop=True)
df
```

```python
indices = pd.Series(df.index, index=df['Song Title']).drop_duplicates()
```

```python
indices = indices.drop_duplicates()
indices
```

```python
from sklearn.metrics.pairwise import linear_kernel
```

```python
def recommend(title, n=10):
    if title not in indices:
        return 'Song not found'
    idx = indices[title]
    sim_score = linear_kernel(tfidf_matrix[idx], tfidf_matrix).flatten()
    similar_idx = sim_score.argsort()[::-1][1:n+1]

    return pd.DataFrame(df['Song Title'].iloc[similar_idx])
```

```python
recommend('Koi Haseena')
```
